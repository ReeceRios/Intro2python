```python
import pandas as pd
from transformers import pipeline
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
```

```python
data = """Sounds great! Can't wait
Looking forward to seeing you 
Sorry, wish I could make it 
Unfortunately I'll be out of town that week 
Hey Reece! I love, love but I can't attend 
Count me in! 
My best to the groom and bride, I will not be able to make it
I will not join
Thank you for the invite, I'm excited to celebrate
Can I bring a plus 1? I'd love to join """.split("\n")

data
```

```python
df = pd.DataFrame() 
for d in data:
    row = pd.DataFrame(classifier(d)[0]).set_index('label').T
    row['text'] = d
    df = pd.concat([df, row])
df = df.reset_index(drop=True)
df
```
